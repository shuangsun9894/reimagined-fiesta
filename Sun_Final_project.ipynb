```r
products_df<- read.csv('https://raw.githubusercontent.com/ALY6000/datasets/main/kaggle/summer-products-with-rating-and-performance_2020-08.csv')
```

Q1:Is retail price positively correlated with price?
 
The retail price and the price are positively correlated. From the piontplot graph that is the trend with price and retail price, we can see that there is a positive trend that means with the increase of the retail price, the price of the selling price will increase accordingly.


```r
plot(products_df$price,products_df$retail_price,
ylab="retial_price",xlab="price",main="plot5:is retail price positively
correlated with price",col="cadetblue4")
```

Q2:Does more badges are associated with higher units sold?

The high badges are related to higher units sold. From the boxplot graph with the badges and units sold, we can see that compared with the mean vaule of the unites soldand the badges count, the average units sold are higher with more badges count.

```r
boxplot(products_df$units_sold~products_df$badges_count,ylab="units_sold",xlab ="1:badges_count",main="does more badges is
associated with higher units sold",col="red",las=1 )
```

Q3:Does urgency banner is associated with higher units sold?

The urgency banner is not really related to higher units sold.I used pointplot at this time, and the units sold are exactely the same with 0 urgency banner and with 1 urgency banner. There are factors that may affect the results, for example, if the price is too high or the size of the clothes is too big, even with urgency banner, the units sold might not be affected.


```r
plot(products_df$units_sold~products_df$has_urgency_banner,ylab="units_sold",xlab ="1:has_urgency_banner",main="does urgency banner is associated with higher units sold",col="black",las=1 )
```

Q4: Does higher rating associated with higher units sold?

The higher rating is associated with higher units sold. According to the pointplot graph, it could be observed that with higher ratings, the products units sold are higher than the one with lower ratings. The trend is positive and upward.


```r
plot(products_df$rating,products_df$units_sold,ylab="unit_sold",xlab="rating",main="plot8:does higher rating
associated with higher units sold",col="pink")
```

Q5: Does merchant rating associated with higher units sold?

The merchant rating is associated with higher units sold, similar to the flow of the rating one. According to the pointplot graph, it could be observed that with higher merchant ratings, the products units sold are higher than the one with lower merchant ratings. The trend is positive and upward. And it can be observed that there are more units sold with merchant rating than the rating. Therefore, merchant rating is more important than the product ratings.

```r
plot(products_df$merchant_rating,products_df$units_sold, ylab="units_sold",xlab="merchant_rating",main="plot9:does merchant
rating associated with higher units sold",col="black")
```

```r
summarize_df<- function(dataset) 
 {
  n = ncol(dataset)
  name = names(dataset)
  for (i in 1:n) {
    if (length(unique(dataset[,i])) == nrow(dataset)) {
      cat(name[i], "unique identification. 5 example values are ", paste0(dataset[1:5,i],collapse=", "), '"\n', sep = "")
    } else if (is.numeric(dataset[,i])) {
      # integer
      if (all(dataset[,i]==as.integer(dataset[,i]))) {
        cat(name[i], min(dataset[,i]), " and ",
            max(dataset[,i]), " are its min and max values, with ",
            names(sort(table(dataset[,i])))[1], 
            " being the most common value\n",'"\n',  sep = "")
      } else { # double
        cat(name[i], ": ", "its mean and interquartile range are ",
            mean(dataset[,i]), " and ", IQR(dataset[,i]),
            ", respectively\n", sep = "")
      }
    }
  }
}


```

```r
summarize_df(df)
```
